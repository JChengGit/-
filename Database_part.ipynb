```python
# 唐诗

import json,pymysql,os,re
from langconv import *
from openccpy.opencc import *

CONN = pymysql.connect(host='127.0.0.1', port=3306, user='root', passwd='1234', db='poetry')
cur = CONN.cursor()

rootdir = 'G:\PoetryAnalysis\data\shi'
files = os.listdir(rootdir)

for file in files:
    if re.match('poet', file):
        path = os.path.join(rootdir,file)
        f = open(path,'rb')
        shi = json.loads(f.read())

        for s in shi:
            # title = Opencc.to_simple(Converter('zh-hans').convert(s['title']))
            # author = Opencc.to_simple(Converter('zh-hans').convert(s['author']))
            title = s['title']
            author = s['author']
            content = ''
            for i in s['paragraphs']:
                # content += Opencc.to_simple(Converter('zh-hans').convert(i))
                content += i
            sql = "INSERT INTO cht(title,author,content) VALUES('%s','%s','%s')"%(title,author,content)

            try:
                cur.execute(sql)
            except Exception as e:
                print(e)
                print(file)
                print(author)
                print(title,len(title))
                print(content,len(content))
                

CONN.commit()
CONN.close()

print('Fin')
    
```

```python
# 宋词

import json,pymysql,os,re
from langconv import *

CONN = pymysql.connect(host='127.0.0.1', port=3306, user='root', passwd='1234', db='poetry')
cur = CONN.cursor()

rootdir = 'G:\PoetryAnalysis\data\ci'
files = os.listdir(rootdir)

for file in files:
    if re.match('ci.song', file):
        path = os.path.join(rootdir,file)
        f = open(path,'rb')
        shi = json.loads(f.read())

        for s in shi:
            title = S2T(s['rhythmic'])
            author = S2T(s['author'])
            content = ''
            for i in s['paragraphs']:
                content += S2T(i)
            sql = "INSERT INTO cht(title,author,content) VALUES('%s','%s','%s')"%(title,author,content)

            try:
                cur.execute(sql)
            except Exception as e:
                print(e)
                print(file)
                print(author)
                print(title,len(title))
                print(content,len(content))
                

CONN.commit()
CONN.close()

print('Fin')
    
```

```python
from langconv import *

def T2S(sentence):
    sentence = Converter('zh-hans').convert(sentence)
    return sentence

def S2T(sentence):
    sentence = Converter('zh-hant').convert(sentence)
    return sentence

if __name__=="__main__":
    traditional_sentence = '铁'#'憂郁的臺灣烏龜'
    simplified_sentence = S2T(traditional_sentence)
    print(simplified_sentence)
```

```python
# -*- coding: utf-8 -*-
# 从数据库转txt出来训练词向量

import pymysql
CONN = pymysql.connect(host='127.0.0.1', port=3306, user='root', passwd='1234', db='poetry')
cur = CONN.cursor()

f = open('cht_poetry_corpus.txt','a+',encoding='utf-8')
for i in range(1,350000):
    try:
        sql = 'select content from cht where id =%s;'%i
        cur.execute(sql)
        q = cur.fetchall()[0][0]
        p = ' '.join(q) + '\n'
        f.write(p)
    except Exception as e:
        print(p)
        print(e)
        break
f.close()
print('Fin')
```
