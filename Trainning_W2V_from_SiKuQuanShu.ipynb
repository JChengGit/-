```python
import os, time
from gensim.models import word2vec

corpus = 'E:/材料/四库全书/'
saveplace = 'E:/材料/四库全书_按字分词/SiKuQuanShu_Zi.txt'
```

```python
# 读单本，分字，返回list
def readbook2list(path):
    lst = os.listdir(path)
    lines = []
    for title in lst:
        if title != '目录.txt':
            lines.append(' '.join(title).split('.')[0])
            with open(path+title,'r') as txt:
                article = txt.readlines()
            for i in article:
                lines += ' '.join(i).replace('\n','').split('。')
                
    return lines
```

```python
# 从四库全书中读取文本，分词

content = []
for i in os.listdir(corpus):
    if not os.path.isfile(corpus+i):
        content.append(i)
total = str(len(content))
count = 0
with open(saveplace,'a+',encoding='utf-8') as txt:
    for book in content:
        count += 1
        countstr = '[ '+str(count)+' / '+total+' ] '
        print(time.strftime('%Y-%m-%d %H:%M:%S ')+countstr+book)
        lines = readbook2list(corpus+book+'/')
        for line in lines:
            if len(line)>2:
                txt.write(line)
```

```python
sentences = word2vec.LineSentence(saveplace)
model = word2vec.Word2Vec(sentences, min_count=10)
model.save('E:/材料/四库全书_按字分词/Zi_vec.bin')
print('Finish')
```
